```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('Mall_Customers (1).csv')
```

```python
dataset
```

```python
X = dataset.iloc[:, [3,4]].values
```

```python
X
```

```python
from sklearn.cluster import DBSCAN

# Assuming X is your dataset
aff =  DBSCAN(eps=0.5,min_samples=5, metric='euclidean', metric_params=None, algorithm='auto', leaf_size=30, p=None, n_jobs=None)
y_aff = aff.fit_predict(X)
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised['cluster_group']=y_aff
```

```python
supervised
```
